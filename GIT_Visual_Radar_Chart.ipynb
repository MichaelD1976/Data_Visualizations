## Utilizing the Soccerplot library for graphical analytics
### Creating a Radar Chart for player data comparison
##### Ilkay Gundogan and Kevin De Bruyne were both instrumental in Manchester City's midfield for 2022-23
##### We compare them using a range of parameters taken from statsbomb data and standardized to per 90 minutes  

```python
# pip install soccerplot

import pandas as pd
from soccerplots.radar_chart import Radar
pd.set_option('display.max_columns', None)
pd.options.mode.chained_assignment = None  
```

```python
# read in csv file of Premier league player data taken from statsbomb
df = pd.read_csv('C:\\Users\\Documents\\Python Outputs\\Prem_RATINGS_COMB.csv')
```

```python
df.head()
```

```python
# Select our players for comparison
df = df[(df['Name'] == 'Kevin De Bruyne') | (df['Name'] =='Ilkay Gundogan')].reset_index()
```

```python
df
```

```python
# Header details
df.columns
```

```python
# Create new columns for required comparison parameters based on per 90 mins

df['Recoveries'] = df['Recov'] / df['90s']
df['Fouls_Made'] = df['Fls_Made'] / df['90s']
df['Fouls_Won'] = df['Fls_Won'] / df['90s']
df['Attack_Take-ons'] = df['At_Tk-on'] / df['90s']
df['Prog_Carries'] = df['PrgC'] / df['90s']
df['Pass_Compltd'] = df['Pass_Cmp'] / df['90s']
df['Key_Passes'] = df['KP'] / df['90s']
df['Pass_to_Final_1/3'] = df['Pass_1/3'] / df['90s']
df['non_pen_XG'] = df['npxG'] / df['90s']
df['Prog_Pass_Dst'] = df['PassPgDst'] / df['90s']
df['Tackles_Won'] = df['TklW'] / df['90s']
df['X_Assists'] = df['xA'] / df['90s']
df['Touches'] = df['Touches'] / df['90s']
```

```python
# Select chosen columns for our comparison
df = df[['Name',
         'Recoveries',
         'Fouls_Made',
         'Fouls_Won',
         'Attack_Take-ons',
         'Prog_Carries',
         'Pass_Compltd',
         'Key_Passes',
         'Pass_to_Final_1/3',
         'non_pen_XG', 
         'Prog_Pass_Dst',
         'Tackles_Won',
         'Sh/90',
         'X_Assists',
         'Touches']]
```

```python
df.head()
```

```python
# get player comparison parameters as a list
params = list(df.columns)
params = params[1:] # reads the column headersstarting after player name
params
```

```python
# add ranges to list (min & max) of tuple pairs, add/subtract approx 25% for improved final display
# create empty lists to fill
ranges = []
a_values = []
b_values = []

for x in params:
    a = min(df[params][x]) # finds min value of each parameter
    a = a- (a*.25)         # gives our minimum value (25% less than actual)
    b = max(df[params][x]) # finds max value of each parameter
    b = b + (b*.25)        # gives our maximum (25% more than actual)

    ranges.append((a,b))  # append to list

for x in range(len(df['Name'])):
    if df['Name'][x] == 'Kevin De Bruyne':
        a_values = df.iloc[x].values.tolist()
    if df['Name'][x] == 'Ilkay Gundogan':
        b_values = df.iloc[x].values.tolist()

#remove the name parameters, so only comparison parameters are stored
a_values = a_values[1:]
b_values = b_values[1:]
# create a new list with these two lists inside
values = [a_values,b_values]
```

```python
# Radar chart title info and format 

title = dict(
    title_name = 'Kevin De Bruyne',
    title_color = 'red',
    subtitle_name = 'Man City',
    subtitle_color = 'red',
    title_name_2 = 'Ilkay Gundogan',
    title_color_2 = 'blue',
    subtitle_name_2 = 'Man City',
    subtitle_color_2 = 'blue',  
    title_fontsize = 18,
    subtitle_fontsize = 15
    )
```

```python
radar = Radar()

fig,ax = radar.plot_radar(ranges=ranges,params=params,values=values,
                          radar_color=['red','blue'],
                          alphas=[.65,.5], # transparent colors
                          title=title,
                          compare=True)
```
